```python
%matplotlib inline
import matplotlib.pyplot as plt
from scipy.io import wavfile as wav
from scipy.fftpack import fft
from scipy.interpolate import interp1d
import numpy as np
rate, data = wav.read('1_esben_1s/chunk23.wav')
fft_out = np.abs(fft(data))
#plt.plot(data, np.abs(fft_out))
#plt.show()

print(fft_out)


N = rate # Number of samplepoints
yf = fft_out
xf = range(0, round(N/2))

yf_human = fft_out[100:400]
xf_human = range(100, 400)
cubic_interploation_model = interp1d(xf_human, yf_human, kind = "cubic")

X_ = np.linspace(100, 399, 1000)
Y_=cubic_interploation_model(X_)

plt.figure(figsize=(20, 10))
plt.subplot(311)
plt.plot(xf, yf[0:round((N/2))])
plt.subplot(312)
plt.plot(xf_human, yf_human)
plt.subplot(313)
plt.plot(X_, Y_) #plt.show()

```
